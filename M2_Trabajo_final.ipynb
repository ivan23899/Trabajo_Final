### 7. Analisis del tempo de las canciones

Utilizar el comando `print(f"...")` donde corresponda para responder a las siguientes preguntas

Para los artistas con una popularidad mayor a 90 y cuyas canciones han sido lanzadas despues del 2015, encontrar la siguiente informacion:

- ¬øCu√°l es la canci√≥n m√°s r√°pida en t√©rminos de tempo y en qu√© a√±o fue lanzada?
- ¬øCu√°l es la canci√≥n m√°s lenta en t√©rminos de tempo y en qu√© a√±o fue lanzada?
- ¬øGraficar el numero de canciones que fueron lanzadas por a√±o y identificar cu√°l fue a√±o con m√°s canciones?


```python
# tu c√≥digo aqu√≠ 
# ¬øCu√°l es la canci√≥n m√°s r√°pida en t√©rminos de tempo y en qu√© a√±o fue lanzada?
artistas_populares_despues_2015 = df_spotify[(df_spotify['popularity']>90)&(df_spotify['year']>2015)]
ordenado_por_tempo_rapido = artistas_populares_despues_2015.sort_values(by='tempo', ascending=False)
cancion_mas_rapida = ordenado_por_tempo_rapido.iloc[0]
print(f"La cancion mas rapida es {cancion_mas_rapida['name']} con un tempo de {cancion_mas_rapida['tempo']}, fue lanzada en {cancion_mas_rapida['year']}")
```

```python
# tu c√≥digo aqu√≠ 
# ¬øCu√°l es la canci√≥n m√°s lenta en t√©rminos de tempo y en qu√© a√±o fue lanzada?
ordenado_por_tempo_rapido = artistas_populares_despues_2015.sort_values(by='tempo', ascending=True)
cancion_mas_lenta = ordenado_por_tempo_rapido.iloc[0]
print(f"La cancion mas rapida es {cancion_mas_lenta['name']} con un tempo de {cancion_mas_lenta['tempo']}, fue lanzada en {cancion_mas_lenta['year']}")
```

```python
# tu c√≥digo aqu√≠ 
# ¬øGraficar el numero de canciones que fueron lanzadas por a√±o y identificar cu√°l fue a√±o con m√°s canciones?
import matplotlib.pyplot as plt

conteo_por_anio = artistas_populares_despues_2015['year'].value_counts().sort_index()
anio_mas_canciones = conteo_por_anio.idxmax()
maximo_canciones = conteo_por_anio.max()

plt.figure(figsize=(12, 6))
plt.bar(conteo_por_anio.index, conteo_por_anio.values, color='skyblue')
plt.xlabel('A√±o de Lanzamiento')
plt.ylabel('N√∫mero de Canciones')
plt.title('N√∫mero de Canciones Lanzadas por A√±o')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.annotate(f'M√°s Canciones: {maximo_canciones}', xy=(anio_mas_canciones, maximo_canciones), xytext=(anio_mas_canciones + 5, maximo_canciones + 10),
             arrowprops=dict(arrowstyle='->', lw=1.5), fontsize=12, color='red')

plt.tight_layout()
plt.show()

print(f"El a√±o con m√°s canciones fue {anio_mas_canciones} con un total de {maximo_canciones} canciones.")





```

### 12. Crear un playlist con las canciones mas populares

Crea una `lista` de reproducci√≥n con la canci√≥n m√°s popular de cada a√±o desde 1980 hasta 2020.

El resultado debe ser de tipo List con 41 valores de tipo str (una canci√≥n para cada a√±o)
p.ej. ['cancion1', 'cancion2', 'cancion3', ...]
NOTA: En caso de empates, elige la canci√≥n con el valor m√°s alto de energy.

```python
# tu c√≥digo aqu√≠

df_filtrado = df_spotify[(df_spotify['year'] >= 1980) & (df_spotify['year'] <= 2020)]
canciones_mas_populares = []
for year in range(1980, 2021):
    canciones_del_anio = df_filtrado[df_filtrado['year'] == year]
    cancion_mas_popular_ano = canciones_del_anio.sort_values(by=['popularity', 'energy'], ascending=[False, False]).iloc[0]
    nombre_cancion_mas_popular = cancion_mas_popular_ano['name']
    canciones_mas_populares.append(nombre_cancion_mas_popular)
for i, cancion in enumerate(canciones_mas_populares, start=1980):
    print(f"A√±o {i}: {cancion}")
```
